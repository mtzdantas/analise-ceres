### Limpeza de Dados referentes as informações pessoais de Discentes

```python
import pandas as pd
```

```python
discentes_info = pd.read_csv('../../data/raw/dados-pessoais-discentes.csv', sep=';')
discentes_info.head()
```

```python
# Filtrando os dados para incluir apenas os discentes do curso de Sistemas de Informação
discentes_info = discentes_info[(discentes_info['curso'] == 'SISTEMAS DE INFORMAÇÃO')]
discentes_info.head()
```

```python
# Removendo colunas desnecessárias
discentes_info = discentes_info.drop(['id_discente', 'ano_nascimento', 'estado_origem', 'cidade_origem', 'estado', 'nivel_ensino', 'forma_ingresso', 'periodo_ingresso', 'status',	'cotista',	'tipo_cota',	'descricao_tipo_cota'], axis=1)
```

```python
# Exibindo as informações do DataFrame
discentes_info.info()
```

```python
# Preenchendo valores ausentes com "Não Informado"
discentes_info = discentes_info.fillna("Não Informado")
```

```python
discentes_info
```

```python
# Exportando o DataFrame filtrado para um novo arquivo CSV
discentes_info.to_csv('../../data/processed/discentes_info.csv', index=False)
```
