### Limpeza de Dados referentes aos discentes egressos

```python
import pandas as pd
```

```python
anos = [2020, 2021, 2022, 2023]
egressos_por_ano = {}
dfs = []
```

```python
for ano in anos:
    caminho = f'../../data/raw/egressos/discentes-egressos-{ano}.csv'
    df = pd.read_csv(caminho, sep=';')

    # Filtra por id_unidade_gestora (1482 é o CERES) e nivel_ensino
    df = df[
        (df['id_unidade_gestora'] == 1482) &
        (df['nivel_ensino'] == 'GRADUAÇÃO')
    ]

    # Seoara bacharelado e licenciatura de Geografia
    df.loc[df['id_curso'] == 2000044, 'nome_curso'] = 'BAC. GEOGRAFIA'
    df.loc[df['id_curso'] == 111635055, 'nome_curso'] = 'LIC. GEOGRAFIA'

    # Remove colunas desnecessárias
    df = df.drop([
        'periodo_conclusao', 'periodo_ingresso', 'modalidade_educacao',
        'forma_ingresso', 'tipo_discente', 'nivel_ensino',
        'id_unidade', 'nome_unidade'
    ], axis=1)

    # Armazena o DataFrame filtrado no dicionário
    egressos_por_ano[ano] = df
```

```python
# Juntando os DataFrames de todos os anos em um único
for ano, df in egressos_por_ano.items():
    df['ano'] = ano
    dfs.append(df)

egressos_2020_2023 = pd.concat(dfs, ignore_index=True)
```
