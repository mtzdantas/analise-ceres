### Limpeza de Dados referentes a Avaliação da Docência

```python
import pandas as pd
```

```python
docentes_dct = pd.read_csv('../../data/raw/docentes.csv', sep=';')
avaliacao_docentes = pd.read_csv('../../data/raw/avaliacaodocencia.csv', sep=';')

docentes_dct.head()
```

```python
# Filtrando apenas docentes do Departamento de Computação e Tecnologia
docentes_dct = docentes_dct[docentes_dct['lotacao'] == 'DEPARTAMENTO DE COMPUTAÇÃO E TECNOLOGIA']
docentes_dct
```

```python
avaliacao_docentes.head()
```

```python
# Filtrando avaliações dos docentes do Departamento de Computação e Tecnologia
avaliacao_dct = avaliacao_docentes[avaliacao_docentes['nome_docente'].isin(docentes_dct['nome'])]
avaliacao_dct
```

```python
# Conferindo valores nulos e tipos de dados
avaliacao_dct.info()
```

```python
# Removendo colunas desnecessárias para nosso estudo
avaliacao_dct = avaliacao_dct.drop(['id_turma', 'ano', 'periodo', 'autoavaliacao_aluno_media', 'autoavaliacao_aluno_DP'], axis=1)
```

```python
avaliacao_dct
```

```python
# Exportando o DataFrame filtrado para um novo arquivo CSV
avaliacao_dct.to_csv('../../data/processed/avaliacoes_dct.csv', index=False)
```
